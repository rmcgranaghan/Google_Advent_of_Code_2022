```python
txt_file = '/Users/ryanmcgranaghan/Documents/Google_Advent_of_Code_2022/day1_input.txt'
f = open(txt_file,"r")
txt_data = f.read().split('\n')#.remove('')
f.close()

# int_data = [int(x) for x in txt_data]
# int_data
```

```python
elves_cals = []
loop_sum = 0
for t in txt_data:
    if t == '':
        elves_cals.append(loop_sum)
        loop_sum = 0
    else:
        loop_sum += int(t)
print('the elf with the most calories is {}'.format( np.where( elves_cals == np.max(elves_cals))[0][0] + 1 ))

print('the elf with the most calories has {} calories'.format( np.max(elves_cals) ))
```

```python
sum_top_three = 0
elves_cals_popped = elves_cals.copy()

for i in range(0,3):
    idx_most_cals = np.where( elves_cals_popped == np.max(elves_cals_popped))[0][0]
    print(idx_most_cals)
    sum_top_three += elves_cals_popped[idx_most_cals]
    elves_cals_popped.pop(idx_most_cals)

print('the answer to Day 1 Part 2 is {}'.format(sum_top_three))
```
